```python
txt = """TURMA A
ana (12, 13, 15, 12, 13, 15, 14);
joao (9,7,3,6,9);
xico (12,16).
TURMA B
ana (12, 13, 15, 12, 13, 15, 14); 
joao (9,7,3,6,9,12);
xico (12,16).
"""
```

```python
%pip install lark
from lark import Lark,Token
from lark.tree import pydot__tree_to_png
from lark import Transformer
from lark import Discard
import datetime

grammar = '''
//Regras Sintaticas
start: turma+
turma : INFO alunos 
alunos:
      | aluno (PV aluno)+ PF
aluno: ID "(" notas ")"
notas: NUMERO ("," NUMERO)*
//Regras Lexicográficas
ID: WORD
INFO: /TURMA [A-Z]/
NUMERO: INT
PV:";"
PF:"."
//Tratamento dos espaços em branco
%import common.WS
%import common.NEWLINE
%import common.INT
%import common.WORD
%ignore NEWLINE
%ignore WS
'''

def query(nome,nota,turma):
  data = datetime.datetime.now()
  with open('resultado.sql','a') as f:
    f.write(f"""INSERT INTO Resultado ("nomeAluno", "nota", "dataInsercao", "turma") 
                          VALUES  ("{nome}",{nota},{data},"{turma}");\n""")
  f.close()


def trataElementos(elemento,lst,dic_notas,self):
  for elem in elemento[1]:
      media = 0
      for x in elem[1]:
        media += x

        query(elem[0],x,elemento[0])

        if str(x) in dic_notas.keys():
          dic_notas[str(x)].update({elem[0]})
        else:
          dic_notas[str(x)] = {elem[0]}

      media /= self.num_disciplinas
      lst.append((elem[0],elem[1],int(media)))
  return lst


class Transformer(Transformer):
  count = 0
  dic_turmas = {}
  num_disciplinas = 0
  def start(self,elementos):
    return [self.count,self.dic_turmas]
    
  def turma(self,elemento):
    dic_notas = {}
    lst = []
    lst = trataElementos(elemento,lst,dic_notas,self)


    self.dic_turmas[elemento[0]] = {"turma": lst, "notas": dic_notas, "num_disciplinas": self.num_disciplinas}
    self.num_disciplinas = 0
    pass

  def alunos(self,elemento):
    return elemento
  
  def aluno(self,elemento):
    self.count+=1
    return (str(elemento[0]), elemento[1])
    
  def notas(self,elemento):
    if len(elemento) > self.num_disciplinas:
      self.num_disciplinas = len(elemento)
    return list(elemento)

  def INFO(self,info):
    return str(info)
  
  def NUMERO(self,numero):
    return int(numero)

  def PV(self,pf):
    return Discard
  
  def PF(self,pf):
    return Discard

  pass

f = open('resultado.sql','w')
f2 = open('resultado.md','w')
f3 = open('resultado.html','w')
f.close()

p = Lark(grammar) 


tree = p.parse(txt)
#print(tree.pretty())
#for element in tree.children:
  #print(element)

def markdwon(data,f):
  f.write("# Visualizador de turmas\n")
  for turma in data[1].keys():
    alunos_ = []
    f.write(f"## {turma}\n")
    f.write("### Lista de alunos \n")
    for aluno in data[1][turma]["turma"]:
      alunos_.append((aluno[0],aluno[2]))
      f.write(f"- {aluno[0]}\n")

    f.write("### Notas \n\n")
    f.write("| Aluno | Media |\n")
    for aluno in alunos_:
      f.write(f"| {aluno[0]} | {aluno[1]} |\n")
    f.write("\n")
  f.close()
  

def header_table(len,f,turma):
  f.write(f"<h1>{turma}<code><br /></code></h1>")
  f.write("""<table style="border-collapse: collapse; width: 100%; height: 81px;" border="1">
  <tbody>""")
  f.write("""<tr style="height: 18px;">""")
  f.write("""<td style="width: 10%; height: 18px; text-align: center;">Nome</td>""")
  for i in range(len):
    f.write(f"""<td style="width: 10%; height: 18px; text-align: center;">Nota{i+1}</td>""")
  f.write("""<td style="width: 10%; height: 18px; text-align: center;">M&eacute;dia</td>""")
  f.write("""<td style="width: 10%; height: 18px; text-align: center;">&nbsp;</td>""")
  f.write("""</tr>""")

def html(data,f):
  for turma in data[1].keys():
    header_table(data[1][turma]["num_disciplinas"],f,turma)

    for aluno in data[1][turma]["turma"]:
      f.write("""<tr style="height: 21px;">""")
      f.write(f"""<td style="width: 10%; text-align: center; height: 21px;">{aluno[0]}</td>""")
      for i in range(data[1][turma]["num_disciplinas"]):
        nota = aluno[1][i] if i < len(aluno[1]) else "-"
        f.write(f"""<td style="width: 10%; text-align: center; height: 21px;">{nota}</td>""")

      f.write(f"""<td style="width: 10%; text-align: center; height: 21px;">{aluno[2]}</td>""")

      if aluno[2] > 10:
        f.write("""<td style="width: 10%; text-align: center; height: 21px;"><img src="https://html-online.com/editor/tiny4_9_11/plugins/emoticons/img/smiley-embarassed.gif" alt="embarassed" /></td>""")
      else:
        f.write("""<td style="width: 10%; text-align: center; height: 21px;"><img src="https://html-online.com/editor/tiny4_9_11/plugins/emoticons/img/smiley-cry.gif" alt="cry" /></td>""")
      f.write("""</tr>""")
    f.write("""</tbody>""")
    f.write("</table>")
  f.close()
  
  pass

data = Transformer().transform(tree) # chamar o transformer para obter
markdwon(data,f2)
html(data,f3)

print(data)
f2.close()
f3.close()
#print(data)
```
